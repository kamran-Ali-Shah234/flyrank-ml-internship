# Week 2 – ML Task Framing

## 1. My Lane as an ML Task

### Selected Lane
Lane 2 – Refresh / Content Opportunity Scoring

### ML Task Type
Ranking / Scoring (Supervised Machine Learning)

### Why this task?

The goal is not simply to classify pages as "good" or "bad". Instead, the objective is to rank pages according to how urgently they should be reviewed for content refresh. A supervised machine learning model can learn patterns from historical search performance and produce a priority score for each page.

## 2. Target or Proxy

For the starter dataset I will use:

**Target (Proxy):**

is_declining_label

This proxy identifies whether a page is considered declining based on the starter dataset.

In a future capstone project, a stronger target would be:

Features from the previous 90 days → predict decline during the next 30 days.

Using a future outcome reduces the risk of data leakage.

## 3. Success Metric

The main success metric is:

Precision@50

Reason:

The SEO team usually reviews only a limited number of pages. Precision@50 measures how many of the top 50 recommended pages truly need review.

Additional metrics include:

- ROC-AUC
- Average Precision
- Precision@20

## 5. Why ML Beats a Fixed Rule

A fixed rule might only consider one or two signals, such as impressions or content age.

Machine learning can learn relationships among many observable features including:

- Search volume
- Impressions
- CTR
- Average position
- Content age
- Engagement
- Sessions

The starter notebooks showed that a machine learning model achieved higher Precision@50 than the hand-written baseline rule. This suggests that combining multiple observable signals can improve page prioritization.

The model should be viewed as decision support rather than an automatic decision maker. Human review remains essential.
