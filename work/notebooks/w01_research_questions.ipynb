## 2. Research question

Which pages should be reviewed first by an SEO or content team.

- Refresh outdated content

- Expand thin content

### Cost of a Wrong Recommendation
If the model recommends the wrong pages:

- Time may be spent improving pages with little opportunity.
- High-value pages needing attention may be ignored.
- SEO resources could be used inefficiently.

### Why Machine Learning?

Many search signals interact together (impressions, clicks, CTR, age, engagement, position, etc.). Machine learning may identify useful patterns that are difficult to capture using a single hand-written scoring rule.

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)

print("\nRows:", len(df))
print("Columns:", len(df.columns))

print("\nAverage Search Volume:")
print(round(df["search_volume"].mean(),2))

print("\nAverage Impressions:")
print(round(df["impressions_90d"].mean(),2))

print("\nAverage CTR:")
print(round(df["ctr"].mean(),4))

print("\nAverage Position:")
print(round(df["avg_position"].mean(),2))

print("\nCorrelation between Search Volume and Impressions:")
print(round(df["search_volume"].corr(df["impressions_90d"]),3))
```
